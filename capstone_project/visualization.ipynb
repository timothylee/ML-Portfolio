```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline

act_train = pd.read_csv('c:/ml/redhat/input/act_train.csv', parse_dates=['date'])
act_test = pd.read_csv('c:/ml/redhat/input/act_test.csv', parse_dates=['date'])
people = pd.read_csv('c:/ml/redhat/input/people.csv', parse_dates=['date'])

df_train = pd.merge(act_train, people, on='people_id')
df_test = pd.merge(act_test, people, on='people_id')
del act_train, act_test, people
```

```python
date_x = pd.DataFrame()
date_x['Class probability'] = df_train.groupby('date_x')['outcome'].mean()
date_x['Frequency'] = df_train.groupby('date_x')['outcome'].size()

plt = date_x.plot(
    secondary_y='Frequency', 
    figsize=(30, 10), 
    grid=True, 
    kind='area', 
    fontsize=20, 
    colormap='summer'
)

plt.set_ylabel('Probability of Positive Outcome', fontsize=20)
plt.right_ax.set_ylabel('Frequency of Activity', fontsize=20)
plt.set_xlabel('Date of Activity', fontsize=20)
plt.set_title('Date vs Frequency of Activity/Outcome Probability', fontsize=20)
```
